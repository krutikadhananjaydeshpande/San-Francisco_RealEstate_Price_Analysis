```python
counties = ['https://www.redfin.com/county/303/CA/Alameda-County',
'https://www.redfin.com/county/309/CA/Contra-Costa-County',
'https://www.redfin.com/county/323/CA/Marin-County',
'https://www.redfin.com/county/330/CA/Napa-County',
'https://www.redfin.com/county/340/CA/San-Francisco-County',
'https://www.redfin.com/city/17490/CA/San-Mateo',
'https://www.redfin.com/county/345/CA/Santa-Clara-County',
'https://www.redfin.com/county/350/CA/Solano-County',
'https://www.redfin.com/county/351/CA/Sonoma-County']

county_page_links =[]
for county in counties:
    county_page_links.append(county)
    for i in range(2,5):
        county_page_links.append(county+'/page-'+str(i))
print(len(county_page_links))        
print(county_page_links)  
```

```python
properties= defaultdict(set) 

for url in county_page_links:    
    county_name = re.findall('CA/([^/]+)',url) # key       
    text = requests.get(url, headers=req_headers).text
    soup = BS(text,'html.parser')     
    county_text = soup.find('div',attrs={'id':'results-display'}) 
    all_links = [a['href'] for a in county_text('a') if a.has_attr('href')] 
    regex = r"^/CA" 
    for link in all_links:
        if re.match(regex,link):
            mlink = "https://www.redfin.com" + link
            properties[county_name[0]].add(mlink)

print(len(properties))
print(properties)
```

```python
# Creating empty lists for required parameters
#State = []
County =[]
City = []
Street= []
Zipcode = []
Property_type = []
Status = []
Price = []
Year_built = []
Beds = []
Bath = []
Sqft = []
Acre=[]
Lot_area = []
HOA_due = []
School_score = []
Walk_score = []
Bike_score = []
Transit_score = []
Garage_availibility = []
Redfin_estimate = []
property_type_value = np.nan
year_built_value = np.nan
lot_area_value = np.nan

# Loop through each property in the 'properties' dictionary
for property in properties:
    
    # Loop through each property URL in the current property set
    for prop in properties[property]:  
        # Send an HTTP GET request to the property URL and retrieve the HTML content
        html_text = requests.get(prop, headers=req_headers).text
        # Parse the HTML content using BeautifulSoup
        soup = BS(html_text, 'html.parser')
        
        try:
            # Find the 'span' with the class 'table-label' and text 'County'
            county_label = soup.find('span', class_='table-label', text='County')
            # Get the following 'div' with the class 'table-value'
            county_value = county_label.find_next('div', class_='table-value').text.strip()
            # Append the county value to the 'County' list
            County.append(county_value)
            print(county_value)
        except:
            # If the value is not found, assign NaN to 'County'
            County.append(np.nan)
        
        # Extract State information
        try:
            # Split and extract the state from the 'dp-subtext' div
            state_value = soup.find('div', attrs={'class': 'dp-subtext'}).text.split(', ')[1].split(' ')[0]
        except:
            state_value = np.nan
        # Append the state value to the 'State' list
        State.append(state_value)
        
        # Extract City information
        try:
            # Split and extract the city from the 'dp-subtext' div
            city_value = soup.find('div', attrs={'class': 'dp-subtext'}).text.split(', ')[0]
        except:
            city_value = np.nan
        # Append the city value to the 'City' list
        City.append(city_value)
        
        # Extract Street information
        try:
            # Split and extract the street from the 'street-address' div
            street_value = soup.find('div', attrs={'class': 'street-address'}).text.split(',')[0]
        except:
            street_value = np.nan
        # Append the street value to the 'Street' list
        Street.append(street_value)
        
        # Extract Zipcode information
        try:
            # Split and extract the zipcode from the 'dp-subtext' div
            zipcode_value = soup.find('div', attrs={'class': 'dp-subtext'}).text.split(', ')[-1].split(' ')[-1]
        except:
            zipcode_value = np.nan
        # Append the zipcode value to the 'Zipcode' list
        Zipcode.append(zipcode_value)
        
        # Extract Property Type information
        try:
            # Find all 'div' elements with class 'keyDetail font-weight-roman font-size-base'
            property_value = soup.findAll('div', attrs={'class': 'keyDetail font-weight-roman font-size-base'})
            for tag in property_value:
                # If the 'span' element inside the div has text 'Property Type'
                if tag.span.text == 'Property Type':
                    # Extract and store the property type
                    property_type_value = tag.find('span', attrs={'class': 'content text-right'}).text
        except:
            # If not found, assign NaN to 'Property_type'
            property_type_value = np.nan
        # Append the property type value to the 'Property_type' list
        Property_type.append(property_type_value)
        
        # Extract Status information
        try:
            # Find all 'div' elements with class 'keyDetail font-weight-roman font-size-base'
            status_text = soup.findAll('div', attrs={'class': 'keyDetail font-weight-roman font-size-base'})
            status = ''
            for tag in status_text:
                # If the 'span' element inside the div has text 'Status'
                if tag.span.text == 'Status':
                    # Check if the status has a 'DefinitionFlyoutLink' or a simple 'content text-right'
                    status = tag.find('span', attrs={'class': 'DefinitionFlyoutLink inline-block underline clickable'})
                    if status is None:
                        status = tag.find('span', attrs={'class': 'content text-right'})
        except:
            # If not found, assign NaN to 'Status'
            status = np.nan
        # Append the status value to the 'Status' list
        Status.append(status)
        
        # Extract Price information
        try:
            # Extract the price value after splitting by '$'
            price_value = soup.find('div', attrs={'class': 'statsValue'}).text.split('$')[1]
        except:
            # If not found, assign NaN to 'Price'
            price_value = np.nan
        # Append the price value to the 'Price' list
        Price.append(price_value)
        
        # Extract Year Built information
        try:
            # Find all 'div' elements with class 'keyDetail font-weight-roman font-size-base'
            year_built_text = soup.findAll('div', attrs={'class': 'keyDetail font-weight-roman font-size-base'})
            for tag in year_built_text:
                # If the 'span' element inside the div has text 'Year Built'
                if tag.span.text == 'Year Built':
                    # Extract and convert the year built to an integer
                    year_built_value = int(tag.find('span', attrs={'class': 'content text-right'}).text)
        except:
            # If not found, assign NaN to 'Year_built'
            year_built_value = np.nan
        # Append the year built value to the 'Year_built' list
        Year_built.append(year_built_value)
        
        # Extract Bed information
        try:
            # Extract the bed value by splitting by 'B'
            bed_value_str = soup.find('div', attrs={'data-rf-test-id': 'abp-beds'}).text.split('B')[0]
            # Convert the bed value to an integer
            bed_value = int(bed_value_str)
        except:
            # If not found, assign NaN to 'Beds'
            bed_value = np.nan
        # Append the bed value to the 'Beds' list
        Beds.append(bed_value)
        
        # Extract Bath information
        try:
            # Extract the bath value by splitting by 'B'
            bath_value_str = soup.find('div', attrs={'data-rf-test-id': 'abp-baths'}).text.split('B')[0]
            # Convert the bath value to a floating-point number
            bath_value = float(bath_value_str)
        except:
            # If not found, assign NaN to 'Bath'
            bath_value = np.nan
        # Append the bath value to the 'Bath' list
        Bath.append(bath_value)
        
        # Extract Sqft information
        try:
            # Extract the square footage value, remove commas, and convert to a floating-point number
            sqft_value_str = soup.find('div', attrs={'data-rf-test-id': 'abp-sqFt'}).text.split('S')[0].replace(',', '')
            sqft_value = float(sqft_value_str)
        except:
            # If not found, assign NaN to 'Sqft'
            sqft_value = np.nan
        # Append the square footage value to the 'Sqft' list
        Sqft.append(sqft_value)
        
        # Extract HOA Due information
        try:
            # Find all 'div' elements with class 'keyDetail font-weight-roman font-size-base'
            hoa_text = soup.findAll('div', attrs={'class': 'keyDetail font-weight-roman font-size-base'})
            hoa_due_value = ''
            for tag in hoa_text:
                # If the 'span' element inside the div has text 'HOA Dues'
                if tag.span.text == 'HOA Dues':
                    # Extract and store the HOA due value
                    hoa_due_value = tag.find('span', attrs={'class': 'content text-right'}).text
        except:
            # If not found, assign NaN to 'HOA_due'
            hoa_due_value = np.nan
        # Append the HOA due value to the 'HOA_due' list
        HOA_due.append(hoa_due_value)
        
        # Extract School Score information
        try:
            # Initialize an empty list to store school ratings
            school = []
            # Find the 'div' element with class 'schools-content'
            school_soup = soup.find('div', attrs={'class': 'schools-content'})
            for tag in school_soup.findAll('span', attrs={'class': 'rating-num font-size-base font-weight-bold'}):
                # Append each school rating to the 'school' list
                school.append(tag.text)
            # Convert school ratings to integers
            school_rating = [int(i) for i in school]
        except:
            # If not found, assign NaN to 'School_score'
            school_rating = np.nan
        # Calculate the mean of school ratings and round it to 2 decimal places
        School_score.append(np.round(np.mean(school_rating), 2))
        
        # Extract Walk Score information
        try:
            # Find the 'div' element with class 'transport-icon-and-percentage walkscore'
            walk_score = soup.find('div', attrs={'class': 'transport-icon-and-percentage walkscore'}).text
        except:
            # If not found, assign NaN to 'Walk_score'
            walk_score = np.nan
        # Append the walk score value to the 'Walk_score' list
        Walk_score.append(walk_score)
        
        # Extract Transit Score information
        try:
            # Find the 'div' element with class 'transport-icon-and-percentage transitscore'
            transit_score = soup.find('div', attrs={'class': 'transport-icon-and-percentage transitscore'}).text
        except:
            # If not found, assign NaN to 'Transit_score'
            transit_score = np.nan
        # Append the transit score value to the 'Transit_score' list
        Transit_score.append(transit_score)
        
        # Extract Bike Score information
        try:
            # Find the 'div' element with class 'transport-icon-and-percentage bikescore'
            bike_score = soup.find('div', attrs={'class': 'transport-icon-and-percentage bikescore'}).text
        except:
            # If not found, assign NaN to 'Bike_score'
            bike_score = np.nan
        # Append the bike score value to the 'Bike_score' list
        Bike_score.append(bike_score)
        
        # Extract Garage Availability information
        try:
            # Find all 'span' elements with class 'entryItemContent'
            garage_avail = soup.findAll('span', attrs={'class': 'entryItemContent'})
            for tag in garage_avail:
                # Check if the text contains "Garage"
                if "Garage" in tag.text:
                    value = 'Yes'
                    break
                else:
                    value = 'No'
        except:
            # If not found, assign NaN to 'Garage_availibility'
            value = np.nan
        # Append the garage availability value to the 'Garage_availibility' list
        Garage_availibility.append(value)
        
        # Extract Redfin Estimate information
        try:
            # Find all 'div' elements with class 'keyDetail font-weight-roman font-size-base'
            redfin_estimate_text = soup.findAll('div', attrs={'class': 'keyDetail font-weight-roman font-size-base'})
            redfin_estimate_value = ''
            for tag in redfin_estimate_text:
                # If the 'span' element inside the div has text 'Redfin Estimate'
                if tag.span.text == 'Redfin Estimate':
                    # Extract and split the Redfin Estimate value and get the part after '$'
                    redfin_estimate_value = tag.find('span', attrs={'class': 'content text-left'}).text.split('$')[1]
        except:
            # If not found, assign NaN to 'Redfin_estimate'
            redfin_estimate_value = np.nan
        # Append the Redfin Estimate value to the 'Redfin_estimate' list
        Redfin_estimate.append(redfin_estimate_value)
```
